```python
import pandas as pd
import matplotlib.pyplot as  plt 
plt.rcParams['font.sans-serif'] = ["WenQuanYi Zen Hei"]
plt.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_excel(r"附件/IT行业收入表.xlsx")
X = df[['工龄']]
Y = df['薪水']

plt.scatter(X,Y)
plt.xlabel('年龄')
plt.ylabel('薪水')
```

```python
from sklearn.linear_model import LinearRegression
regr = LinearRegression()
regr.fit(X,Y)
```

```python
plt.scatter(X,Y)
plt.plot(X,regr.predict(X),color='red')
plt.xlabel('年龄')
plt.ylabel('薪水')
```

```python
print('系数a:'+ str(regr.coef_[0]))
print('截距b:' + str(regr.intercept_))
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_regr = PolynomialFeatures(degree=2)
X_ = poly_regr.fit_transform(X)

regr = LinearRegression()
regr.fit(X_,Y)
plt.scatter(X,Y)
plt.plot(X,regr.predict(X_),color='blue')
plt.grid()
```

```python
import statsmodels.api as sm
X2 = sm.add_constant(X_)
est = sm.OLS(Y,X2).fit()
est.summary()
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(Y,regr.predict(X_))
print('R2 score: %.2f' % r2)
```
